```python
import json
```

```python
# Let's write a function to load data
def load_data(filename) :
    with open(filename , 'r') as f:
        data = json.load(f)
        return data
```

```python
data = load_data("data.json")
```

```python
# write a function to display user and their connections
def display_users(data):
    print("Users and their connection\n")
    for user in data['users']:
        print(f" ID:{user['id']} - {user['name']} is friends with : {user['friends']} and liked pages are {user['liked_pages']}")
    print("\npages Information \n")
    for pages in data['pages']:
        print(f"{pages['id']}: {pages['name']} ")

display_users(data)
```
